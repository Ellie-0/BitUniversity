# 3.0 - Now Try This:
- Use the functional method of programming to build a two subplot figures ```ax``` and ```ax1``` respectively.
- First create a fig by callting the ```plt.figure``` function, set the ```figsize``` to 20,10.
- Create the subplots.
**Hint**: ```fig.add_subplot(something,something,something)```.
- plot ```df['AAPL']``` for ```ax```.
- plot ```df['SPX']``` for ```ax```.
- add tick labels for both ax and ax2
    - Name them in the sequenct ```['Phase I','Phase II',...,'PHASE VII']```
    - both should have a ```rotation``` = 45.
- set title for ```ax``` and ```ax1``` by using ```set_title```. Title should be the name of your stock index.
- set xlabel for both charts by using ```set_xlabel``` to 'Stages'


## Answer 

`````python
fig = plt.figure(figsize = (20,10))
ax = fig.add_subplot(1,2,1)
ax1 = fig.add_subplot(1,2,2)
ax.plot(df['AAPL'])
ax1.plot(df['SPX'])

ax.set_xticklabels(['Phase I','Phase II', 'Phase III','Phase IV','PHASE V','PHASE VI','PHASE VII'],
                           rotation = 45)

ax1.set_xticklabels(['Phase I','Phase II', 'Phase III','Phase IV','PHASE V','PHASE VI','PHASE VII'],
                           rotation = 45)

ax.set_title('STOCK INDEX FOR APPLE')
ax1.set_title('STOCK INDEX FOR SPX')

ax.set_xlabel('Stages')
ax1.set_xlabel('Stages')
`````
